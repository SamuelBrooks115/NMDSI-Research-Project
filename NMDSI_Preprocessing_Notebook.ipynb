```python
"""2 columns that represent paid amount and recieved amount for each transaction,
following code determines if that is necessary. Only necessary if there is a transaction fee
or if there is a transaction between two currencies. (EX: US Dollar -> Euro)
"""


# Print the column names to verify
print("Dataset columns:", dataset.columns)

# Determine if there are differences in Amount Received and Amount Paid
print('Amount Received = to Amount Paid')
amount_difference = not dataset['Amount Received'].equals(dataset['Amount Paid'])
print(f"Is there a difference between Amount Received and Amount Paid? {amount_difference}")

# Determine if there are differences in Receiving Currency and Payment Currency
print('Receiving Currency = to Payment Currency')
currency_difference = not dataset['Receiving Currency'].equals(dataset['Payment Currency'])
print(f"Is there a difference between Receiving Currency and Payment Currency? {currency_difference}")

# Count and summarize discrepancies
transactions_with_fees = dataset[dataset['Amount Received'] != dataset['Amount Paid']].shape[0]
transactions_with_currency_conversion = dataset[dataset['Receiving Currency'] != dataset['Payment Currency']].shape[0]

print(f"Number of transactions with potential fees: {transactions_with_fees}")
print(f"Number of transactions with currency conversions: {transactions_with_currency_conversion}")
```

```python
# Create a flag to identify transactions with the same payment and receiving currency
dataset['Same Currency'] = dataset['Payment Currency'] == dataset['Receiving Currency']

# Filter the transactions where the currencies are the same
same_currency_transactions = dataset[dataset['Same Currency']]

# Further filter where the Amount Paid and Amount Received are different
different_amount_transactions = same_currency_transactions[same_currency_transactions['Amount Paid'] != same_currency_transactions['Amount Received']]

# Count the number of such transactions
count_different_amount = different_amount_transactions.shape[0]

print(f"Number of transactions with the same payment and receiving currency but different amounts: {count_different_amount}")

#Print List of Recieving and Payment Currencies
print(sorted(dataset['Receiving Currency'].unique()))
print(sorted(dataset['Payment Currency'].unique()))
```

```python
# Filter transactions with different currencies
different_currency_transactions = dataset[dataset['Payment Currency'] != dataset['Receiving Currency']]

# Calculate the implied exchange rate safely using .loc
different_currency_transactions.loc[:, 'Implied Exchange Rate'] = (
    different_currency_transactions['Amount Received'] / different_currency_transactions['Amount Paid']
)

# Group by currency pairs and calculate the average implied exchange rate
currency_pairs = different_currency_transactions.groupby(
    ['Payment Currency', 'Receiving Currency']
)['Implied Exchange Rate'].mean().reset_index()

# Rename columns for clarity
currency_pairs.columns = ['From Currency', 'To Currency', 'Average Exchange Rate']

# Display the results
print(currency_pairs)

#Save the results to a CSV file for further use
currency_pairs.to_csv('implied_exchange_rates.csv', index=False)



```

```python
# Load original dataset
dataset = pd.read_csv('HI-Small_Trans.csv')

# Load the implied exchange rates from the CSV file
exchange_rates = pd.read_csv('implied_exchange_rates.csv')

# Ensure consistent column naming
exchange_rates.columns = ['From Currency', 'To Currency', 'Average Exchange Rate']


def calculate_expected_amount(row):
  """
  Calculate the expected amount recieved based on the implied exchange rate.

  First checks to see if convsersion is needed, if the payment and recieving currency are the same
  then it is not needed. If not the function uses the implied exchange rate to calculate the expected amount.

  Parameters:
  --------
  - row : pd.Series
    A pandas series representing a single transaction. Required to have the following columns:
      - 'Payment Currency' - currency of the payment
      - 'Receiving Currency' - currency recieved
      - 'Amount Paid' - amount paid in payment currency
      - 'Amount Received' - amount recieved in receiving currency

  - exchange_rates : pd.DataFrame
    A pandas DataFrame containing the implied exchange rates for each currency pair. Required to have the following columns:
      - 'From Currency' - currency of the payment
      - 'To Currency' - currency recieved
      - 'Average Exchange Rate' - implied exchange rate

  Returns:
  --------
  float or None
    - The expected amount recieved based on the implied exchange rate.
  """
  if row['Payment Currency'] == row['Receiving Currency']:
        # No conversion needed, expected amount is the same as paid amount
        return row['Amount Paid']
  else:
        # Look up the exchange rate
        rate = exchange_rates[
            (exchange_rates['From Currency'] == row['Payment Currency']) &
            (exchange_rates['To Currency'] == row['Receiving Currency'])
        ]['Average Exchange Rate'].values
        # If a rate is found, calculate the expected amount
        if len(rate) > 0:
            return row['Amount Paid'] * rate[0]
        else:
            # If no rate is found, we can't calculate the expected amount
            return None


# Apply the function to calculate the expected received amount
dataset['Expected Received Amount'] = dataset.apply(calculate_expected_amount, axis=1)

# Calculate the difference between the expected and actual received amount
dataset['Amount Difference'] = dataset['Expected Received Amount'] - dataset['Amount Received']

# Define a threshold for what constitutes a significant difference (indicating potential fees)
threshold = 1  # Adjust as needed

# Flag transactions with significant differences
dataset['Has Transaction Fee'] = dataset['Amount Difference'].apply(lambda x: abs(x) > threshold)

# Count the number of transactions with significant differences
count_significant_difference = dataset['Has Transaction Fee'].sum()

print(f"Number of transactions with significant differences (potential fees): {count_significant_difference}")

# Display the transactions with potential fees
transactions_with_fees = dataset[dataset['Has Transaction Fee']]

# Display relevant columns for analysis
print(transactions_with_fees[['Timestamp', 'From Bank', 'Account', 'To Bank', 'Account.1',
                              'Amount Paid', 'Payment Currency', 'Amount Received', 'Receiving Currency',
                              'Expected Received Amount', 'Amount Difference', 'Has Transaction Fee']])

# Save the result to a CSV file for further analysis
transactions_with_fees.to_csv('transactions_with_potential_fees.csv', index=False)
```

```python
'''
This is to get more information from the transactions with fees

Looks at the transaction types associated with the transactions with fees and determines if they are flagged as laundering or not.
'''

#Load the CSV file
file_path = 'transactions_with_potential_fees.csv'
dataset = pd.read_csv(file_path)

# Group by 'Payment Format' and calculate the required sums
summary = dataset.groupby('Payment Format').agg(
    Sum_of_Trans_W_Fees=('Payment Format', 'size'),  # Count of transactions with fees
    Num_Laundering=('Is Laundering', 'sum')          # Sum of laundering transactions
).reset_index()

# Rename the columns
summary.columns = ['Transaction Type', 'Sum of Trans W Fees', 'Num Laundering']

#Spits out a table with the transaction type with fees the # of transacitons and the sum of the laundering cases within them
print(summary)
```

```python
import datetime
import os
from typing import Callable, Optional
import pandas as pd
from sklearn import preprocessing
import numpy as np
import torch
import gdown


# Function to download the file
def download_file():
  # Extract the file ID and set the destination file name
  file_id = '1l2KLgqTKFbcAQdk6yWRk7A_8YUSKfF1y'  # File ID
  destination = 'HI-Small_Trans.csv'  # File Name

  # Construct the download URL and download the file
  download_url = f'https://drive.google.com/uc?id={file_id}'
  gdown.download(download_url, destination, quiet=False)

# Load the dataset
file_name = 'HI-Small_Trans.csv'
if os.path.exists(file_name):
    dataset = pd.read_csv(file_name)
else:
    download_file()
    dataset = pd.read_csv(file_name)

# Define the preprocessing functions
def label_encoder(dataset, columns):
    le = preprocessing.LabelEncoder()
    for column in columns:
        dataset[column] = le.fit_transform(dataset[column])
    return dataset


def combine_acct_label(dataset):
    '''Concatenate From/To Bank and Account/.1 separated by an '_' (bank_account)'''
    # Create Unique Account IDs
    dataset['Account'] = dataset['From Bank'].astype(str) + '_' + dataset['Account']
    dataset['Account.1'] = dataset['To Bank'].astype(str) + '_' + dataset['Account.1']
    return dataset

def normalize_features(dataset):
    scaler = preprocessing.StandardScaler()
    features = dataset[['Amount Paid', 'Amount Received']]
    scaled_features = scaler.fit_transform(features)
    dataset[['Amount Paid', 'Amount Received']] = scaled_features
    return dataset

def preprocess_data(dataset):
    '''Combine Previous functions and sort and encode currency'''
    dataset = combine_acct_label(dataset)
    dataset = normalize_features(dataset)
    dataset = label_encoder(dataset, ['Payment Format', 'Payment Currency', 'Receiving Currency', 'Account', 'Account.1'])
    currency_lst = sorted(dataset['Receiving Currency'].unique())
    return dataset, currency_lst

# Apply the preprocessing
processed_dataset, currency_lst = preprocess_data(dataset)
print(f'Processed Dataset:\n{processed_dataset.head()}')
#print(f'Currency List:\n{currency_lst}')


# Save the processed dataset to Google Drive
from google.colab import drive
drive.mount('/content/drive')

file_path = '/content/drive/My Drive/processed_dataset.csv'
processed_dataset.to_csv(file_path, index=False)

```
